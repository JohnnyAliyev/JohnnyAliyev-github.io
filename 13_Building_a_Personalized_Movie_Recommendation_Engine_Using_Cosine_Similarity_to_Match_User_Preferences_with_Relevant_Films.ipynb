# *1.Import Libraires*

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

#

# *2.Import Dataset*

```python
movies_data=pd.read_csv('/Users/maegbug/Desktop/ARTIFICIAL INTELLIGENT/Sample Datas for ML/movies.csv')
movies_data.head(16)
```

#

```python
features=['genres','keywords','tagline','cast','director']
features
```

```python
#replacing null values
for i in features:
    movies_data[i]=movies_data[i].fillna('')
movies_data.isnull().sum()
```

#

# *4.Data Preparing*

```python
#combine
c_features= movies_data['genres']+ movies_data['keywords']+movies_data['tagline']+movies_data['cast']+movies_data['director']
pd.DataFrame(c_features)
```

```python
# CONVERTING textual data to numerical data
vectorizer=TfidfVectorizer()
f_vectors=vectorizer.fit_transform(c_features)
print(f_vectors)
```

#

# *5. Finding silimarity*

```python
similarity=cosine_similarity(f_vectors)
# Convert to DataFrame for better readability
df_similarity = pd.DataFrame(similarity, columns=[f'Doc {i}' for i in range(f_vectors.shape[0])],
                             index=[f'Doc {i}' for i in range(f_vectors.shape[0])])

df_similarity = pd.DataFrame(similarity, columns=movies_data['title'], index=movies_data['title'])
pd.options.display.float_format = '{:.1f}'.format
df_similarity['The Dark Knight Rises'].sort_values(ascending=False)
```

```python
movies_data['title'].head()
```

```python
movie_name = input(' Enter your favourite movie name : ')

list_of_all_titles = movies_data['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)

close_match = find_close_match[0]

index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_the_movie]))

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```
